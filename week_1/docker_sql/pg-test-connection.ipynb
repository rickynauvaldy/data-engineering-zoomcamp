````python
query = """
SELECT 
    date(lpep_pickup_datetime) lpep_pickup_date,
    sum(case when passenger_count = 2 then 1 else 0 end) two,
    sum(case when passenger_count = 3 then 1 else 0 end) three
FROM green_taxi_data
group by date(lpep_pickup_datetime)
order by lpep_pickup_date
"""

pd.read_sql(query, con=engine)
````
